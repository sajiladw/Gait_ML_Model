```
import numpy as np
import pandas as pd


Data = pd.read_csv(
    "/home/sdw/PycharmProjects/Lily_RBD_Survival/PD_Control_ML_Fulldata.csv"
)

print(Data["Outcome"].value_counts())

time_points = [3, 4, 5, 6, 7, 8, 9]
times = time_points

N = len(Data)


events_list = []
pct_list = []

for t in times:

    events = (
        (Data["Outcome"] == 1) &
        (Data["FU_duration"] <= t)
    ).sum()

    pct = events / N

    events_list.append(events)
    pct_list.append(pct)



columns = [f"{y} year" for y in time_points]

events_row = pd.Series(events_list, index=columns, name="Events")

pct_row = pd.Series(
    [f"{x:.5f}" for x in pct_list],
    index=columns,
    name="Percentage"
)

table = pd.DataFrame([events_row, pct_row])

print(table)

table.to_csv("cumulative_events_table.csv")
```
